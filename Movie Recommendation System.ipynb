```python
movies_title=movies_title[[0,1]]
movies_title.columns=['S.No.','Name of the Movie']
```

```python
dr=pd.merge(dr,movies_title)
```

```python
ratings=pd.DataFrame(dr.groupby('Name of the Movie').mean()['Ratings'])
```

```python
ratings.head()
```

```python
dr.tail()
ratings['num of ratings']=pd.DataFrame(dr.groupby('Name of the Movie').count()['Ratings'])
```

```python
dr.head()
```

```python
moviemat=dr.pivot_table(index='Item ID',columns='Name of the Movie',values='Ratings')
```

```python
moviemat.head()
```

```python
Toy_Story_Ratings=moviemat['Toy Story (1995)']
```

```python
Toy_Story_Ratings.head(20)
```

```python
similar_to_toy_story=moviemat.corrwith(Toy_Story_Ratings)
```

```python
similar_to_toy_story
```

```python
corr_toystory=pd.DataFrame(similar_to_toy_story, columns=['Correlation'])
```

```python
corr_toystory.dropna(inplace=True)
```

```python
def predict_movies(movie_name):
    movie_user_ratings=moviemat[movie_name]
    similar_to_movie=moviemat.corrwith(movie_user_ratings)
    corr_movie=pd.DataFrame(similar_to_movie,columns=['Correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie=corr_movie.join(ratings['num of ratings'])
    
    predictions=corr_movie[corr_movie['num of ratings']>100].sort_values('Correlation',ascending=False)
    
    return predictions
    
```

```python
predict_my_movie=predict_movies("Toy Story (1995)")
```
